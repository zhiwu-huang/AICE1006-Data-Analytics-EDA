## Manifold Learning Exercises and Solutions

### Exercise 1: Visualizing High-Dimensional Data Using t-SNE


1. Load the `digits` dataset from `sklearn.datasets`.
2. Perform dimensionality reduction using t-SNE to 2 dimensions.
3. Create a scatter plot of the resulting 2D data, coloring points by their digit labels.
    

```python

from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Step 1: Load the digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Step 2: Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Step 3: Scatter plot of t-SNE results
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Digit Class')
plt.title('t-SNE Visualization of Digits Dataset')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.grid()
plt.show()
    
```

### Exercise 2: Using Isomap for Dimensionality Reduction


1. Use the `swiss_roll` dataset from `sklearn.datasets.make_swiss_roll`.
2. Apply Isomap to reduce the dataset's dimensionality to 2 components.
3. Visualize the reduced data using a scatter plot, coloring points by their position along the roll.
    

```python

from sklearn.datasets import make_swiss_roll
from sklearn.manifold import Isomap

# Step 1: Load the swiss roll dataset
X_swiss, color = make_swiss_roll(n_samples=1000, noise=0.1, random_state=42)

# Step 2: Apply Isomap
isomap = Isomap(n_components=2)
X_isomap = isomap.fit_transform(X_swiss)

# Step 3: Scatter plot of Isomap results
plt.figure(figsize=(8, 6))
plt.scatter(X_isomap[:, 0], X_isomap[:, 1], c=color, cmap='Spectral', alpha=0.7)
plt.title('Isomap Dimensionality Reduction on Swiss Roll')
plt.xlabel('Isomap Dimension 1')
plt.ylabel('Isomap Dimension 2')
plt.grid()
plt.show()
    
```

### Exercise 3: Exploring Locally Linear Embedding (LLE)


1. Use the `swiss_roll` dataset again for this exercise.
2. Apply Locally Linear Embedding (LLE) to reduce the dataset's dimensionality to 2 components.
3. Visualize the reduced data using a scatter plot, coloring points by their position along the roll.
    

```python

from sklearn.manifold import LocallyLinearEmbedding

# Step 1: Apply Locally Linear Embedding
lle = LocallyLinearEmbedding(n_neighbors=10, n_components=2, random_state=42)
X_lle = lle.fit_transform(X_swiss)

# Step 2: Scatter plot of LLE results
plt.figure(figsize=(8, 6))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=color, cmap='Spectral', alpha=0.7)
plt.title('Locally Linear Embedding on Swiss Roll')
plt.xlabel('LLE Dimension 1')
plt.ylabel('LLE Dimension 2')
plt.grid()
plt.show()
    
```

### Exercise 4: Comparing Manifold Learning Techniques


1. Apply t-SNE, Isomap, and LLE to the `digits` dataset.
2. Create subplots to compare the 2D visualizations from each method side-by-side.
3. Observe the differences in clustering and structure preservation between the methods.
    

```python

# Step 1: Apply manifold learning techniques
X_tsne_digits = TSNE(n_components=2, random_state=42).fit_transform(X)
X_isomap_digits = Isomap(n_components=2).fit_transform(X)
X_lle_digits = LocallyLinearEmbedding(n_neighbors=10, n_components=2, random_state=42).fit_transform(X)

# Step 2: Create subplots for comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)

# t-SNE plot
axes[0].scatter(X_tsne_digits[:, 0], X_tsne_digits[:, 1], c=y, cmap='viridis', alpha=0.7)
axes[0].set_title('t-SNE')

# Isomap plot
axes[1].scatter(X_isomap_digits[:, 0], X_isomap_digits[:, 1], c=y, cmap='viridis', alpha=0.7)
axes[1].set_title('Isomap')

# LLE plot
axes[2].scatter(X_lle_digits[:, 0], X_lle_digits[:, 1], c=y, cmap='viridis', alpha=0.7)
axes[2].set_title('LLE')

# General settings
for ax in axes:
    ax.set_xlabel('Dimension 1')
    ax.set_ylabel('Dimension 2')
    ax.grid()

plt.suptitle('Comparison of Manifold Learning Techniques on Digits Dataset', fontsize=16)
plt.tight_layout()
plt.show()
    
```
