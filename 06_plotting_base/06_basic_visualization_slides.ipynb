## Matplotlib in a nutshell

Matplotlib is the basic python plotting library: https://matplotlib.org

* Generate publication-quality images
* Export results in pdf, svg, png, jpg
* Mostly intended for *static* 2D graphs
* Rather low-level: build everything from scratch


### Getting started

```python
import matplotlib.pyplot as plt # standard way to import matplotlib
import numpy as np # most of the times you also need numpy
```

```python
# We may want to modify the default style options
print(plt.style.available)
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = [15, 8]
plt.rcParams['font.size'] = 20
```

```python
# generate some data to work with
x = np.linspace(0, 10, 100) # 100 evenly spaced points in the numerical interval [0, 10]
y = x**2 # x squared - a parabola!
```

Let us make a simple **line plot**: $y$ against $x$

```python
fig = plt.figure() # Figure object
plt.plot(x, y); # horizontal, vertical
```

### Multiple lines


We can plot multiple curves in the same figure

```python
x = np.linspace(0, 10, 100) # 100 evenly spaced points in range [0, 10]
y1 = 10*x # linear function
y2 = x**2 # quadratic function
y3 = 2*x**2 - 10*x # another quadratic function
```

```python
fig = plt.figure() # Figure object. 
plt.plot(x, y1) 
plt.plot(x, y2)
plt.plot(x, y3);
```

Don't like the default style? Look at the several options available!

### Style options: line color


```python
x = np.arange(11) # 0, 1, 2, ... 10
y1 = 10*x
y2 = x**2
y3 = 2*x**2 - 10*x
```

```python
plt.figure(figsize=(15, 8)) # speficy figure size
plt.plot(x, y1, 'b') # plot in Blue color
plt.plot(x, y2, 'r') # plot in Red color
plt.plot(x, y3, 'k') # plot in blacK color
plt.grid(True) # show a grid
```

A *style option string* is often used to change the appearance of a plotted object, such as the color:

* `b` $\Rightarrow$ Blue color
* `r` $\Rightarrow$ Red color
* `k` $\Rightarrow$ blacK color

Other common color options: `g` = Green, `m` = Magenta, `c` = Cyan, `y` = Yellow


### Style options: line style

```python
plt.figure()
plt.plot(x, y1, 'b--') # plot in Blue color, dashed line
plt.plot(x, y2, 'r-') # plot in Red color, continouos line
plt.plot(x, y3, 'k-.') # plot in blacK color, dashed-dotted line
plt.grid(True) # show/hide grid
```

We can also specify the line type with style options:

* 'b--' $\Rightarrow$ Blue color, dashed line 
* 'r-' $\Rightarrow$ Red color, continuous line 
* 'k-.' $\Rightarrow$ blacK color, dashed-dotted line 

Note: we combine color and line style information in the style option string. Convention taken from MATLAB

### Style options: marker

```python
plt.figure()
plt.plot(x, y1, 'bo-') # plot in Blue color, continuous line, with circles as markers.
plt.plot(x, y2, 'rs--') # plot in Red color, dashed line, square as markers.
plt.plot(x, y3, 'k^-.'); # plot in blacK color, dashed-dotted line
```

We can also specify the marker type: 

* 'bo-' $\Rightarrow$ Blue color, marker circle, continuous line
* 'rs--' $\Rightarrow$ Red color, marker square, dashed line 
* 'k^-.' $\Rightarrow$ blacK color, marker triangle, dashed-dotted line 

### Other style options

```python
plt.figure()
plt.plot(x, y1, 'bo-', linewidth=4, markersize=15) # plot in Blue color, continuous line, with circles as markers.
plt.plot(x, y2, 'rs--', linewidth=4, markersize=15); # plot in Red color, dashed line, square as markers.
```

Certain style options can only be speficied as extra arguments. Here we changed:
* Line width
* Marker size

Again, check out the matplotlib documentation!

### Axes limits, grid, and legend

```python
plt.figure()
plt.plot(x, y1, 'bo-', label='y1') # plot in Blue color, continuous line, with circles as markers. Set line name y1 for legend
plt.plot(x, y2, 'rs--', label='y2') # plot in Red color, dashed line, stars as markers. Set line name y2  for legend
plt.xlabel('Time (s)') # label for the x axis
plt.ylabel('Speed (m/s)') # label for the y axis 
plt.xlim([0, 10]) # range of the x axis
plt.ylim([-20, 100]) # range of the y axis
plt.legend() # show a legend
plt.grid(True) # show a grid
```

* We can specify:
    * *limits* for the horizontal and vertical axes
    * *labels* for the horizontal and vertical axes
    * *labels* for the different plotted objects, to be shown in a *legend*
* We can turn the *gridline* on and off

* ...

### Subplots

Subplots allow showing several *panels* horizontally/vertically aligned. An example adapted from the 
[matplotlib galley](https://matplotlib.org/gallery/index.html):

```python
# Data for plotting
x = np.linspace(0.0, 10*np.pi, 100)
y1 = 2.0*np.cos(x) * np.exp(-x/10) # cosine function modulated by a negative exponential
y2 = np.cos(x) # cosine function
```

```python
# Create two subplots sharing y axis
fig, ax = plt.subplots(2, sharey=True) # 2 panels aligned vertically
fig.suptitle("Damped and undamped oscillations")
ax[0].plot(x, y1, 'ko-')
ax[0].set_ylabel('Damped oscillation')
ax[0].set_ylim([-2.1, 2.1])

ax[1].plot(x, y2, 'r.-')
ax[1].set_ylabel('Undamped oscillation')
ax[1].set_xlabel('Time (s)');
```

Note: the two plots have the same scale on the $y$ axis (option sharey=True)

### Subplots

An example with rows and columns of panels

```python
n_row = 2 # number of rows
n_col = 3 # number of columns


x = np.linspace(0.0, 1.0)
fig, ax = plt.subplots(n_row, n_col, sharex=True, sharey=True, figsize=(25, 10)) # 2x3 subplots with common x and y axis
fig.suptitle("Several plots")
for row_idx in range(n_row):
    for col_idx in range(n_col):
        ax[row_idx][col_idx].plot(x, (row_idx+1) * x**(col_idx+1) )
        ax[row_idx][col_idx].set_title(f"Subplot {row_idx}, {col_idx}")
```

### Commonly used plots

### Line plot

A line plot is used to represent a continuous function $y = f(x)$. It uses *interpolation* to produce a continuous curve.

For instance, $y = \sin(x), \; x \in [0, 2\pi]$

```python
### Line plot
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)

plt.figure()
plt.plot(x, y, 'o-')
plt.title("y = sin(x)");
```

### Scatter plot

A scatter plot is used to display data points in 2D. Like a lineplot, without interpolation.

```python
### Scatter plot

x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
plt.plot(x, y, 'o'); # use circles as markers
```


Can be implemented using the same ``plot`` function, using a *marker* style option such as ``'o'``.

### Scatter plot cont'd

The ``scatter`` functions allows setting different marker color and size for the data points. 

This allows exploring 4 dimensions altogether!

```python
n_points = 20
x = np.random.randn(n_points)
y = np.random.randn(n_points)
z = np.random.randn(n_points)
w = np.random.rand(n_points)

plt.scatter(x, y, c=z, s=w*2000, alpha=0.3, cmap='viridis')
plt.colorbar();  # show color scale
```

### Histogram

A histogram visually describes the *empirical distribution* of one numerical variable. 

* The numerical range of a variable is divided into *bins*.
* For each bin, the number of times the variable is within the bin range is counted
* The bins may be either user-defined or automatically constructed

```python
import numpy as np
temperatures = 20 + 2 * np.random.randn(1000) # temperatures for different days

plt.figure()
plt.hist(temperatures)
plt.xlabel('Temperature $(^oC)$')
plt.ylabel('Count (-)');
```

### Histogram cont'd

A fancier example with multiple histograms and a legend

```python
# fake temperature data for three months
temp_april = np.random.normal(15, 0.5, 1000) # mean, std, n_samples
temp_may = np.random.normal(20, 1.5, 1000)
temp_june = np.random.normal(25, 1, 1000)


kwargs = dict(histtype='stepfilled', alpha=0.3,  bins=40)
plt.hist(temp_april, **kwargs, label='april')
plt.hist(temp_may, **kwargs, label='may')
plt.hist(temp_june, **kwargs, label='june')
plt.xlabel('Temperature $(^oC)$')
plt.ylabel('Count (-)');
plt.legend();
```

### Histogram  in 2D

Histograms may be extended to 2-dimensional point clouds. Use the ``hist2d`` function

```python
mean = [0, 3] #
cov = [[1, 1],
       [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 100000).T

plt.hist2d(x, y, bins=100, cmap='Blues')
plt.colorbar()
plt.title("histogram")
plt.xlabel("x")
plt.ylabel("y");
```

The histogram count is represented using a *color map*.

### Histogram  in 2D cont'd

Histograms may be extended to 2-dimensional point clouds. Use the ``hist2d`` function

```python
mean = [0, 3] #
cov = [[1, 1],
       [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 200000).T

fig, ax = plt.subplots(1, 2, figsize=(35, 8))
fig.suptitle("2D histogram vs. scatterplot")
ax[0].hist2d(x, y, bins=100, cmap='Blues')
ax[0].set_title("2D histogram")
ax[1].scatter(x, y)
ax[1].set_title("scatterplot");
```

The 2D histogram may be more informative than a 2D scatterplot.

### Boxplot

A box plot summarizes the salient aspects of an empirical distribution. 

```python
import numpy as np
data = np.random.randn(1000) # random data point

plt.boxplot(data);
```

* The orange line is the *median* of the distribution
* 50% of the points are inside the box
* Most of the points are expected to be within the interval defined by the *wiskers*
* Only *outliers* are plotted individually


### Boxplot cont'd

It is less informative than a histogram, but perhaps easier to read. Less is (sometimes) more!

```python
data = np.random.randn(1000)

fig, ax = plt.subplots(1, 2)
fig.suptitle("Histogram vs. Boxplot")
ax[0].hist(data, orientation="horizontal")
ax[0].set_title("Histogram")
ax[1].boxplot(data)
ax[1].set_title("Boxplot");
```

### Boxplot cont'd

We often look at several boxplots simultaneously.

```python
temp_april = np.random.normal(15, 0.5, 1000) # mean, std, n_samples
temp_may = np.random.normal(20, 1.5, 1000)
temp_june = np.random.normal(25, 1, 1000)

plt.boxplot([temp_april, temp_may, temp_june], labels = ['april', 'may', 'june'])
plt.ylabel("Temperature ($^oC$)");
```

### Bar plot

A bar plot is often the more intuitive than a scatter/line plot when the quantity on the $x$ axis is categoric (i.e. not numeric)

```python
import matplotlib.pyplot as plt

names = ['apple', 'orange', 'lime']
values = [10, 20, 15]
fig, ax = plt.subplots(1, 3, figsize=(30, 10), sharey=True)
ax[0].set_title("Bar plot")
ax[0].bar(names, values)
ax[0].set_ylabel("Water content (%)")
ax[1].set_title("Scatter plot")
ax[1].scatter(names, values)
ax[1].set_ylabel("Water content (%)")
ax[2].set_title("Line plot")
ax[2].plot(names, values);
ax[2].set_ylabel("Water content (%)");
```

The line plot is particularly misleading in this example: how to interpolate apples and oranges?
